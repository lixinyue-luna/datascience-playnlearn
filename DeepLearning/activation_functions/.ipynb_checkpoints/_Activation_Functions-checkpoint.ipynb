# Activation Functions
Also known as Transfer Functions, Activation Functions are used between neural networks to map resulting value from the previous layer into a desired range to feed into the next layer. An activation function is usually an abstraction representing the rate of action potential firing in the cell ([wiki](https://en.wikipedia.org/wiki/Activation_function)). 

```python
import numpy as np
import matplotlib.pyplot as plt
from activations import *
```

```python
x = np.linspace(-5, 5, 501)
```

```python
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_relu = relu(x)
y_leaky = leaky_relu(x, a=0.5)

d_sigmoid = sigmoid_derivative(x)
d_tanh = tanh_derivative(x)
d_relu = relu_derivative(x)
d_leaky = leaky_relu_derivative(x, a=0.5)
```

```python
xticks = np.linspace(-5, 5, 21)
yticks1 = np.linspace(-2, 5, 15)
yticks2 = np.linspace(-1, 1, 5)

plt.figure(figsize=(16,9), clear=True)

plt.subplot(211)
plt.plot(x, y_sigmoid)
plt.plot(x, y_tanh)
plt.plot(x, y_relu, linewidth=3)
plt.plot(x, y_leaky)
plt.axhline(0, color='black', linewidth=1.5, linestyle='dashed')
plt.axvline(0, color='black', linewidth=1.5, linestyle='dashed')
plt.xticks(xticks)
plt.xlim(-5, 5)
plt.ylim(-2.5, 5)
plt.legend(['sigmoid', 'tanh', 'relu', 'leaky_relu'])
plt.grid(True)
plt.title('Activations')

plt.subplot(212)
plt.plot(x, d_sigmoid)
plt.plot(x, d_tanh)
plt.plot(x, d_relu, linewidth=3)
plt.plot(x, d_leaky)
plt.axvline(0, color='black', linewidth=1.5, linestyle='dashed')
plt.xticks(xticks)
plt.xlim(-5, 5)
plt.ylim(0, 1)
plt.legend(['sigmoid_derivative', 'tanh_derivative', 'relu_derivative', 'leaky_relu_derivative'])
plt.grid(True)
plt.title('Derivatives')

plt.show()
```
