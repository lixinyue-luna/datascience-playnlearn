```python
import numpy as np
import matplotlib.pyplot as plt
```

## Sigmoid Funcion
$sigmoid(x) = \frac{1}{1+e^{-x}}$ is a non-linear function widely used in logistic regression in machine learning and deep learning. 

$x$ could be either a real number, a vector, or a matrix. More general form of the sigmoid function is as below:
$$ \text{For } x \in \mathbb{R}^n \text{, } sigmoid(x) = sigmoid
\begin{pmatrix}
    x_1  \\
    x_2  \\
    ...  \\
    x_n  \\
\end{pmatrix} = 
\begin{pmatrix}
    \frac{1}{1+e^{-x_1}}  \\
    \frac{1}{1+e^{-x_2}}  \\
    ...  \\
    \frac{1}{1+e^{-x_n}}  \\
\end{pmatrix}\tag{1} $$

```python
def sigmoid(x):
    '''
    Compute sigmoid of x
    
    Arguments:
    x: a scalar or vector/matrix represented as numpy array
    
    Return:
    s: sigmoid(x)
    '''
    s = 1/(1 + np.exp(-x))
    return s
```

```python
x = np.arange(1,11,1)
sigmoid(x)
```

## Sigmoid Gradient
In backpropagation, gradients (a.k.a. slope or derivative) of activation functions are used to optimize loss functions. Use $\sigma(x)$ to denote the sigmoid funciton, then its derivative with respect to input x is:
$$\sigma'(x) = \sigma(x) (1 - \sigma(x))  \tag{2}$$

```python
def sigmoid_derivative(x):
    '''
    Compute the derivative of sigmoid(x)
    
    Arguments:
    x: a scalar or vector/matrix represented as numpy array
    
    Return:
    ds: derivative of sigmoid fucntion
    '''
    # first calculate the sigmoid function
    s = sigmoid(x)

    # then use formula (2) to compute the derivative
    ds = s * (1-s)
    return ds
```

```python
sigmoid_derivative(x)
```

```python
# visualize the sigmoid function and its derivative
data = np.linspace(-8, 8, 1000)
s = sigmoid(data)
ds = sigmoid_derivative(data)

xticks = np.linspace(-8, 8, num=33, endpoint=True)
s_yticks = np.arange(0, 1.1, 0.1)
ds_yticks = np.arange(0, 0.3, 0.03)

plt.figure(figsize=(16, 9), clear=True)

plt.subplot(211)
plt.plot(data, s)
plt.title('$sigmoid(x)$')
plt.xticks(xticks)
plt.yticks(s_yticks)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-8, 8)
plt.grid(True)

plt.subplot(212)
plt.plot(data, ds)
plt.title('$sigmoid\_derivative(x)$')
plt.xticks(xticks)
plt.yticks(ds_yticks)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-8, 8)
plt.grid(True)

plt.subplots_adjust(top=0.9, bottom=0.1, left=0.1, right=0.9, hspace=0.2, wspace=0.35)

plt.show()
```
