```python
import numpy as np
import matplotlib.pyplot as plt
```

# Tanh
$tanh(x) = \frac{e^{x} - e^{-x}}{e^{x}+e^{-x}}$, a.k.a. hyperbolic tangent activation function, is like logistic sigmoid function. It is s-shaped (sigmoidal) and ranges from -1 to 1. It is mainly used for classification between 2 classes.

$x$ could be either a real number, a vector, or a matrix. More general form of the tanh function is as below:
$$ \text{For } x \in \mathbb{R}^n \text{, } tanh(x) = tanh
\begin{pmatrix}
    x_1  \\
    x_2  \\
    ...  \\
    x_n  \\
\end{pmatrix} = 
\begin{pmatrix}
    \frac{e^{x_1} - e^{-x_1}}{e^{x_1}+e^{-x_1}}  \\
    \frac{e^{x_2} - e^{-x_2}}{e^{x_2}+e^{-x-2}}  \\
    ...  \\
    \frac{e^{x_n} - e^{-x_3}}{e^{x_3}+e^{-x_3}}  \\
\end{pmatrix}\tag{1} $$

```python
def tanh(x):
    '''
    Compute tanh of x
    
    Arguments:
    x: a scalar or vector/matrix represented as numpy array
    
    Return:
    t: tanh(x)
    '''
    t = (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))
    return t
```

```python
x = np.arange(1,11,1)
tanh(x)
```

## Tanh Gradient
In backpropagation, gradients of activation functions are used to optimize loss functions. Use $\sigma(x)$ to denote the tanh funciton, then its derivative with respect to input x is:
$$\sigma'(x) = 1 - \sigma(x)^2 \tag{2}$$

```python
def tanh_derivative(x):
    '''
    Compute the derivative of tanh(x)
    
    Arguments:
    x: a scalar or vector/matrix represented as numpy array
    
    Return:
    ds: derivative of tanh fucntion
    '''
    # first calculate the sigmoid function
    t = tanh(x)

    # then use formula (2) to compute the derivative
    dt = 1 - t**2
    return dt
```

```python
tanh_derivative(x)
```

```python
# visualize the sigmoid function and its derivative
data = np.linspace(-8, 8, 1000)
t = tanh(data)
dt = tanh_derivative(data)

xticks = np.linspace(-8, 8, 33, endpoint=True)
t_yticks = np.linspace(-1, 1, 11)
dt_yticks = np.linspace(0, 1, 11)

plt.figure(figsize=(16, 9), clear=True)

plt.subplot(211)
plt.plot(data, t)
plt.axhline(0, color='red', linewidth=0.8, linestyle='dashed')
plt.axvline(0, color='red', linewidth=0.8, linestyle='dashed')
plt.title('$tanh(x)$')
plt.xticks(xticks)
plt.yticks(t_yticks)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-8, 8)
plt.grid(True)

plt.subplot(212)
plt.plot(data, dt)
plt.axvline(0, color='red', linewidth=0.8, linestyle='dashed')
plt.title('$tanh\_derivative(x)$')
plt.xticks(xticks)
plt.yticks(dt_yticks)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-8, 8)
plt.grid(True)

plt.subplots_adjust(top=0.9, bottom=0.1, left=0.1, right=0.9, hspace=0.2, wspace=0.35)

plt.show()
```
