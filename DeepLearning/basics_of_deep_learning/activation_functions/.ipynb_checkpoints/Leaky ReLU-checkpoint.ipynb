```python
import numpy as np
import matplotlib.pyplot as plt
```

# Leaky ReLU
Leaky ReLU was proposed to solve the 'dying' ReLU problem. It uses a 'leak' to increase the range of the ReLU function so to range from -infinity to infinity. Both the function and its deriviative are monotonic. 

$$leaky\_relu(x) = 
\begin{cases}
\alpha x, & \text{if $x<0$} \\
x, & \text{if $x\geqslant 0$} \\
\end{cases} \tag{1}$$

$\alpha$ is usually 0.01 in Leaky ReLU. If it is not 0.01, then it is called Randomized ReLU.
$x$ could be either a real number, a vector, or a matrix.

```python
def leaky_relu(x, a=0.01):
    '''
    Compute leaky relu of x
    
    Arguments:
    x: a scalar or vector/matrix represented as numpy array
    a: the 
    
    Return:
    r: relu(x)
    '''
    r = np.where(x >= 0, x, a*x)
    return r
```

```python
x = np.arange(-5,6,1)
leaky_relu(x)
```

## Leaky ReLU Gradient
In backpropagation, gradients of activation functions are used to optimize loss functions. Use $\sigma(x)$ to denote the Leaky ReLU funciton, then its derivative with respect to input x is:
$$\sigma'(x) = 
\begin{cases}
\alpha, & \text{if $x<0$} \\
1, & \text{if $x\geqslant 0$} \\
\end{cases} \tag{2}$$

```python
def leaky_relu_derivative(x, a=0.01):
    '''
    Compute the derivative of relu(x)
    
    Arguments:
    x: a scalar or vector/matrix represented as numpy array
    
    Return:
    ds: derivative of tanh fucntion
    '''
    # then use formula (2) to compute the derivative
    dr = np.where(x >= 0, 1, a)
    return dr
```

```python
leaky_relu_derivative(x)
```

```python
# visualize the sigmoid function and its derivative
data = np.linspace(-8, 8, 1000)
r = leaky_relu(data)
dr = leaky_relu_derivative(data)

xticks = np.linspace(-8, 8, 33)
r_yticks = np.linspace(-8, 8, 9)
dr_yticks = np.linspace(0, 1, 6)

plt.figure(figsize=(16, 9), clear=True)

plt.subplot(211)
plt.plot(data, r)
plt.axhline(0, color='red', linewidth=0.8, linestyle='dashed')
plt.axvline(0, color='red', linewidth=0.8, linestyle='dashed')
plt.title('$leaky\_relu(x)$')
plt.xticks(xticks)
plt.yticks(r_yticks)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-8, 8)
plt.grid(True)

plt.subplot(212)
plt.plot(data, dr)
plt.axvline(0, color='red', linewidth=0.8, linestyle='dashed')
plt.title('$leaky\_relu\_derivative(x)$')
plt.xticks(xticks)
plt.yticks(dr_yticks)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-8, 8)
plt.grid(True)

plt.subplots_adjust(top=0.9, bottom=0.1, left=0.1, right=0.9, hspace=0.2, wspace=0.35)

plt.show()
```
