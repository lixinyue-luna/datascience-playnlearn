# Line Plot
Present observations collected at regular intervals.

```python
# use numpy to create 100 equally spaced data points between 0 and 2
x = np.linspace(0, 2, 100)

# initiate an empty figure with no axes
fig = plt.figure(figsize=(10,6))# creates the figure and axes
plt.plot(x, x, label='linear') # reuse current axes and add lines
plt.plot(x, np.power(x, 2), label='quadratic')
plt.plot(x, np.power(x, 3), label='cubic')

# set labels
plt.xlabel('x')
plt.ylabel('y')

# set title
plt.title('Example')
plt.legend()
plt.show()

# save figure
# plt.savefig('image.png')
```

```python
# use seaborn
# initiate an empty figure with no axes
fig = plt.figure(figsize=(10,6))# creates the figure and axes
sns.lineplot(x, x, label='linear') # reuse current axes and add lines
sns.lineplot(x, np.power(x, 2), label='quadratic')
sns.lineplot(x, np.power(x, 3), label='cubic')

# set labels
plt.xlabel('x')
plt.ylabel('y')

# set title
plt.title('Example')
plt.legend()
plt.show()
```

```python
# use a helper function to plot figures of the same style
def helper_plotter(ax, x, y, params=None):
    '''
    Helper function to plot a graph
    
    Arguments:
    ax: axes to draw to
    x: x values as array
    y: y values as array
    params: dictionary of kwargs to pass to ax.plot
    
    Return:
    artists: list of artists added
    '''
    artists = ax.plot(x, y, **params)
    return artists
    
x = np.linspace(0, 2*np.pi, 500)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots(1, 2, sharey=True, clear=True, figsize=(16, 6))
# fig.set_size_inches(12, 6)
params1 = {'color': 'r', 'markevery': 100, 'marker': 'o'}
params2 = {'color': 'b', 'markevery': 100, 'marker': 'x'}

artists1 = helper_plotter(ax[0], x, y1, params1)
artists2 = helper_plotter(ax[1], x, y2, params2)

# 'stickiness' of certain plotting method (not working with this line plot though)
# ax[0].use_sticky_edges=True

# set titles
ax[0].set_title('$sin(x)$')
ax[1].set_title('$cos(x)$')

# set x and y limits
ax[0].set_xlim(0, 2*np.pi)
ax[0].set_ylim(-1.2, 1.2)
ax[1].set_xlim(0, 2*np.pi)
ax[1].set_ylim(-1.2, 1.2)

# set x and y ticks
ax[0].set_xticks(np.arange(0, 2*np.pi+0.5, 0.5))
ax[0].set_yticks(np.arange(-1, 1.2, 0.2))
ax[1].set_xticks(np.arange(0, 2*np.pi+0.5, 0.5))
ax[1].set_yticks(np.arange(-1, 1.2, 0.2))

# set labels
ax[0].set_xlabel('x')
ax[0].set_ylabel('$sin(x)$')
ax[1].set_xlabel('x')
ax[1].set_ylabel('$cos(x)$')

ax[0].grid(True)
ax[1].grid(True)

# set axes scales
ax[1].set_xscale('linear') # if x > 0, can set scale to 'logit', 'log', 'symlog'

plt.show()
```

# Bar Chart
Present relative quantities for multiple categories.

```python
x = ['Apple', 'Pear', 'Orange']
y = [np.random.randint(0, 50), np.random.randint(0, 50), np.random.randint(0, 50)]
plt.bar(x, y)
plt.show()
```

```python
sns.barplot(x, y)
plt.show()
```

# Histogram
Summarize the distribution of a data sample, or density estimates. 

```python
x = np.random.randn(10000)
plt.hist(x, bins=20)
plt.show()
```

```python
sns.distplot(x, bins=20)
plt.show()
```

# Box and Whisker Plot
Summarize the distribution of a data sample.

```python
x = [np.random.randn(100), 3*np.random.randn(100), 5*np.random.randn(100)]
plt.boxplot(x)
plt.show()
```

```python
sns.boxplot(data=x)
plt.show()
```

# Scatter Plot
 Show the relationship between two paired data samples.

```python
x = np.linspace(-5, 5, 1001)
y = x + np.random.randn(1001)
plt.scatter(x, y)
plt.show()
```

```python
sns.scatterplot(x, y)
plt.show()
```
